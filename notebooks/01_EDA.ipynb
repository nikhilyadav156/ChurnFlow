```python

# Cell 1 - Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../data/raw/telco_churn.csv')
print(df.shape)
print(df.head())
```

```python
# Cell 2 - Basic Info
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
print("\nChurn Distribution:\n", df['Churn'].value_counts())
print("\nChurn %:\n", df['Churn'].value_counts(normalize=True)*100)
```

```python
# Cell 3 - Churn Distribution
plt.figure(figsize=(6,4))
df['Churn'].value_counts().plot(kind='bar',
    color=['#2196F3','#F44336'], edgecolor='black')
plt.title('Churn Distribution')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
# ~26% churn → imbalanced dataset → need SMOTE
```

```python
# Cell 4 - Tenure vs Churn
plt.figure(figsize=(8,4))
sns.histplot(data=df, x='tenure', hue='Churn',
             bins=30, palette=['#2196F3','#F44336'])
plt.title('Tenure Distribution by Churn')
plt.tight_layout()
plt.show()
# Customers who churn tend to have lower tenure
```

```python
# Cell 5 - Monthly Charges vs Churn
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges',
            palette=['#2196F3','#F44336'])
plt.title('Monthly Charges vs Churn')
plt.tight_layout()
plt.show()
# Churned customers pay more monthly
```

```python
# Cell 6 - Contract Type vs Churn
plt.figure(figsize=(8,4))
contract_churn = pd.crosstab(df['Contract'],
                              df['Churn'], normalize='index')*100
contract_churn.plot(kind='bar', color=['#2196F3','#F44336'],
                    edgecolor='black')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Percentage')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
# Month-to-month contracts churn the most
```

```python
# Cell 7 - Correlation Heatmap
plt.figure(figsize=(10,7))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True,
            fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
